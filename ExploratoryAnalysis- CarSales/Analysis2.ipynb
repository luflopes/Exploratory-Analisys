```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Filtra (agrupa) o dataset inteiro e retorna as colunas 'brand' e 'name'.
pgt1 = df.groupby(['brand'], as_index=False).count()[['brand','name']]
# Renomeia  coluna no novo dataset
pgt1 = pgt1.rename(columns={'name':'qtd'})
# Ordena o dataset em ordem decrescente de acordo com os valores da coluna 'qtd'
pgt1 = pgt1.sort_values(by='qtd', ascending=False)
pgt1.head()
```

```python
# Cria a figura (p1) e cria o objeto gráfico contido na figura (ax1)
p1, ax1 = plt.subplots(figsize=(10, 10))
# Objeto gráfico irá receber um grafico da biblioteca seaborn
ax1 = sns.barplot(x=pgt1['qtd'], y=pgt1['brand'], orient='h', palette='Reds_r')
ax1.set_title('\n Veículos Por Marca \n', fontdict={'fontsize':15})
ax1.set_ylabel('\n Marca \n', fontdict={'fontsize':15})
ax1.set_xlabel('\n Número Veículos \n', fontdict={'fontsize':15})
p1.show()
```
