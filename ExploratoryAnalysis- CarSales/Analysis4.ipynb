## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Dataset
clean_data_path = "dataset/autos/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
pgt1 = df.groupby(['brand', 'vehicleType']).mean()
pgt1 = pgt1.reset_index()
temp = pgt1.pivot('brand','vehicleType','price')
```

```python
plt.figure(figsize=(14, 20))
ax = sns.heatmap(data=temp, annot=True, fmt=".0f", cmap="YlGnBu")
ax.set_ylabel('\n Marca \n', fontdict={'fontsize':15})
ax.set_xlabel('\n Tipo de Veículo \n', fontdict={'fontsize':15})
ax.set_title('\n Preço médio de um veículo por marca e por tipo de veículo \n', fontdict={'fontsize':18})
```
